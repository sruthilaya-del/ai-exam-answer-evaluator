```python
import gradio as gr
from google import genai
import os

# Configure Gemini
client = genai.Client(api_key=os.getenv("GEMINI_API_KEY"))

def evaluate_answer(question, model_answer, student_answer, max_marks):

    prompt = f"""
You are a strict Electrical & Electronics Engineering (EEE) exam evaluator.

Follow these rules strictly:
1. Compare semantically with model answer.
2. Identify accuracy level.
3. Detect missing technical keywords.
4. Check formulas, laws, definitions.
5. Deduct marks if incomplete.
6. Do NOT give full marks unless fully correct.
7. Be strict like university external examiner.

Output format must be exactly:

Marks: X / {max_marks}

Accuracy Level:
(Accurate / Partially Correct / Inaccurate)

Missing Keywords:
- keyword 1
- keyword 2
- keyword 3

Comparison Analysis:
(Explain differences clearly.)

Feedback:
(Constructive technical feedback.)

Now evaluate:

Question: {question}

Model Answer: {model_answer}

Student Answer: {student_answer}
"""

    response = client.models.generate_content(
        model="gemini-3-flash-preview",
        contents=prompt
    )

    return response.text


# Gradio Interface
interface = gr.Interface(
    fn=evaluate_answer,
    inputs=[
        gr.Textbox(label="Question"),
        gr.Textbox(label="Model Answer"),
        gr.Textbox(label="Student Answer"),
        gr.Number(label="Maximum Marks")
    ],
    outputs="text",
    title="âš¡ EEE Strict AI Exam Evaluator",
    description="Compares Student Answer with Model Answer and evaluates strictly."
)

interface.launch(share=True)
```
